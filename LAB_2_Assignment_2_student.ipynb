## 2: Plot 12 most dangerous cities

```python
dang_cities = kill.city.value_counts()[0:12]
dang_cities
```

```python
# with seaborn barplot

fig, ax= plt.subplots(figsize=(10,6))
ax=sns.barplot(x=dang_cities.index, y= dang_cities.values)
ax.set_xticklabels(dang_cities.index, rotation=45);
ax.set_title("Most Dangerous Cities", color="blue", fontsize=15);
for p in ax.patches:
   ax.annotate(round(p.get_height()), (p.get_x()+0.5, p.get_height()+0.5))
```

```python
# with seaborn countplot
fig, ax= plt.subplots(figsize=(10,6))
ax=sns.countplot(x="city", data= kill, order=dang_cities.index)
ax.set_xticklabels(dang_cities.index, rotation=45);
ax.set_title("Most Dangerous Cities", color="blue", fontsize=15);
```

## 3: Plot 20 most dangerous states

```python
dang_states= kill.state.value_counts()[:21]
dang_states
```

```python
# with seaborn barplot
fig, ax= plt.subplots(figsize=(10,6))
ax=sns.barplot(x=dang_states.index, y= dang_states.values)
ax.set_xticklabels(dang_states.index, rotation=45);
ax.set_title("Most Dangerous States", color="blue", fontsize=15);
```

```python
# with seaborn countplot
fig, ax= plt.subplots(figsize=(10,6))
ax=sns.countplot(x="state", data= kill, order=dang_states.index)
ax.set_xticklabels(dang_states.index, rotation=45);
ax.set_title("Most Dangerous States", color="blue", fontsize=15);
```
